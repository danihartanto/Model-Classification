```python
#Import scikit-learn dataset library
from sklearn import datasets

#Load dataset
cancer = datasets.load_breast_cancer()
```

```python
# print the names of the 13 features
print("Features: ", cancer.feature_names)

# print the label type of cancer('malignant' 'benign')
print("Labels: ", cancer.target_names)
```

```python
# print data(feature)shape
cancer.data.shape
```

```python
# print the cancer data features (top 5 records)
print(cancer.data[0:5])
```

```python
# print the cancer labels (0:malignant, 1:benign)
print(cancer.target)
```

1. splitting data menggunakan tarin_test_split
2. data training dan testing diambil secara acak. dengan formasi 30% data test dan 70% data train

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3,random_state=12) # 70% training and 30% test
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
akurasi = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:",akurasi)
```

```python
# Model Precision: what percentage of positive tuples are labeled as such?
print("Precision:",metrics.precision_score(y_test, y_pred))

# Model Recall: what percentage of positive tuples are labelled as such?
print("Recall:",metrics.recall_score(y_test, y_pred))
```

1. Validasi menggunakan Cross validation dengan k-fold = 5
2. splitting data dengan cross validation

```python


from sklearn.model_selection import cross_val_score
X = cancer.data
y = cancer.target
clf = svm.SVC(kernel='linear', C=1, random_state=42)
scores = cross_val_score(clf, X, y, cv=5)
scores
```

```python
scores.mean()
```

**Dari perbandingan kedua proses validasi data, maka di dapatkan hasil keakuratan sebagai berikut:**

```python
print('validasi dengan train_test_split  : ',akurasi)
print('validasi dengan cross validation : ',scores.mean())
```
